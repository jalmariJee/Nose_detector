```python
import cv2
import os 
from imutils import paths
from sklearn.model_selection import train_test_split
import random
import re
import glob
```

```python


def convert_to_yolo_format(image):
    
    crop = 5
    x_min = crop
    x_max = image.shape[0]

    y_min = crop
    y_max = image.shape[1]

    image_width = image.shape[0]
    image_height = image.shape[1]
    # Calculate center coordinates
    x_center = (x_min + x_max) / 2 / image_width
    y_center = (y_min + y_max) / 2 / image_height
    
    # Calculate width and height
    width = (x_max - x_min) / image_width
    height = (y_max - y_min) / image_height
    
    return x_center, y_center, width, height



```

```python
#
imagePathRoot = "dataset\\images\\nose"
savePathRoot = "dataset"

images =  paths.list_files(imagePathRoot)

percentageTrain = 0.05
for imgPath in images:
    img = cv2.imread(imgPath)
    nose_yolo = convert_to_yolo_format(img)
    print(imgPath)
    if percentageTrain < random.uniform(0,1):
        pathAppend = "train"
    else:
        pathAppend = "val"
    root = os.path.abspath("") +"\\" +savePathRoot + "\\" + pathAppend + "\\labels"  + "\\" 

    #files = os.listdir(root)
    #for f in files:
     #   os.remove(root + "\\" + f)
    with open(root + re.sub("jpg|png|jpeg", "txt", 
            imgPath, flags=re.IGNORECASE, ).split('\\')[-1], 'w') as file:
        # Append a new line of text
        string = "0 " + str(nose_yolo[0]) + " " + str(nose_yolo[1]) +" "+  str(nose_yolo[2]) +" "+ str(nose_yolo[3])
        file.write(string)
    root =  os.path.abspath("") + "\\" + savePathRoot + "\\" + pathAppend + "\\images"  + "\\" 
    #files = os.listdir(root)
    #for f in files:
    #     os.remove(root + "\\" + f)
    print(root+  re.sub("jpg|png|jpeg", "jpg", imgPath, 
                flags=re.IGNORECASE, ).split('\\')[-1], 'w')
    cv2.imwrite((savePathRoot + "\\" + pathAppend + "\\images"  + "\\" + 
                 re.sub("jpg|png|jpeg", "jpg", imgPath, 
                flags=re.IGNORECASE, ).split('\\')[-1]),img)
    
    
    # Nose
    
    print(f"Box: 0 {nose_yolo[0]} {nose_yolo[1]} {nose_yolo[2]} {nose_yolo[3]}")
```
